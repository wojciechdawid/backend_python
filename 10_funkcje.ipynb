### Czym s domknicia?

Przykad:

### Co to jest dekorator?

Dekorator to, w najprostszych sowach, funkcja, kt贸ra zmienia zachowanie innej funkcji. Robi to, "opakowujc" oryginaln funkcj w dodatkow logik przed lub po jej wykonaniu, bez zmiany jej samej.

```

````python
from functools import wraps

def simple_decorator(function):

    @wraps(function)
    def wrapper(*args, **kwargs):
        print("Co si dzieje przed wywoaniem funkcji.")
        r = function(*args, **kwargs)
        print("Co si dzieje po wywoaniu funkcji.")
        return r

    return wrapper

@simple_decorator
def foo():
    print("To jest foo")

def bar():
    """dokumentacja funkcji bar"""
    print("to jest bar")

@simple_decorator
def add(a: int, b: int) -> int:
    """dodanie"""
    return a + b
    
````

### Dekoratory z parametrami
Dekoratory mog tak偶e przyjmowa parametry, co czyni je jeszcze bardziej elastycznymi. Dekorator przyjmujcy argumenty jest w rzeczywistoci funkcj zwracajc dekorator.

````python
def repeat(num_times):
    def decorator_repeat(function):
        def wrapper(*args, **kwargs):
            for _ in range(num_times):
                result = function(*args, **kwargs)
            return result
        return wrapper
    return decorator_repeat
````

U偶ycie dekoratora z parametrami wymaga dodatkowych nawias贸w:

````python
@repeat(num_times=2)
def greet(name):
    print(f"Hello {name}")
````

### Bonus - dekorator jako klasa:

Tworzenie dekorator贸w przy u偶yciu klas w Pythonie polega na zdefiniowaniu klasy z metod `__call__`, co pozwala jej instancjom zachowywa si jak funkcje. Poni偶ej znajduje si przykad, jak mo偶na przeksztaci dekorator "repeat" w dekorator oparty na klasie:

````python
class Repeat:
    def __init__(self, num_times):
        self.num_times = num_times

    def __call__(self, function):
        # Ta funkcja zostanie wywoana, gdy u偶yjemy instancji klasy jako dekoratora.
        # Zwraca funkcj 'wrapper', kt贸ra opakowuje oryginaln funkcj.
        def wrapper(*args, **kwargs):
            for _ in range(self.num_times):
                result = function(*args, **kwargs)
            return result
        return wrapper

# U偶ycie dekoratora opartego na klasie:
# @Repeat(num_times=3)
def say_hello(name):
    print(f"Hello, {name}!")
````

````python
# r = Repeat(num_times=3)
say_hello = Repeat(num_times=3)(say_hello)

say_hello("World")
````

Gdy dekorator `@Repeat` jest stosowany do funkcji, konstruktor klasy `Repeat` (`__init__`) jest wywoywany, a nastpnie metoda `__call__` klasy jest u偶ywana jako waciwy dekorator dla funkcji `say_hello`. Metoda `__call__` opakowuje oryginaln funkcj w funkcj `wrapper`, zachowujc standardowe zachowanie dekorator贸w.

Dekoratory oparte na klasach mog by szczeg贸lnie u偶yteczne, gdy jest potrzeba przechowywania stanu midzy wywoaniami dekorowanej funkcji lub gdy dekorator musi mie swoje metody opr贸cz `__call__`. Powy偶szy przykad ilustruje, jak dekorator oparty na klasie mo偶e atwo przechowywa stan (w tym przypadku liczb powt贸rze) midzy wywoaniami.
