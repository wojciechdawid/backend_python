````python
start = lambda a: a >= 5
stop = lambda a: a >= 7

def wytnij(lista, *, start, stop):
    if isinstance(lista, str):
        l = list(lista)
    else:
        l = lista
    s1 = False
    for i, v in enumerate(l):
        if isinstance(s1, bool):
            if start(int(v)):
                s1 = i
        else:
            if stop(int(v)):
                s2 = i
                break
    if isinstance(lista, str):
        return ''.join(l[s1:s2])
    else:
        return l[s1:s2]
            
````

````python
def power_factory(i = 1):
    while True:
        yield lambda x: x ** i
        i += 1
````

#### Ćwiczenie kalkulator

````python
import logging
logging.basicConfig(level=logging.INFO, filename='log_kalkulator.log', encoding='utf-8', format='%(asctime)s %(message)s')
logger = logging.getLogger(__name__)

def suma(a, b):
    logger.info("Wykonano dodawanie")
    return a + b

def odj(a, b):
    return a - b

def mul(a, b):
    return a * b

def div(a, b):
    if b == 0:
        logger.error('Zero Division Error')
    else:
        return round(a / b, 2)

def get_value():
    oper = input('Podaj operację (+-*/)')
    a = input('Podaj wartość a: ')
    b = input('Podaj wartość b: ')
    if oper not in ['+', '-', '*', '/']:
        logger.error("Nieprawidłowa operacja")
    else:
        try:
            a = float(a)
            b = float(b)
            return oper, a, b
        except ValueError:
            logger.error('Nieprawidłowe wartości.')

slownik = {
    '+': suma,
    '-': odj,
    '*': mul,
    '/': div
}

def kalkulator():
    oper, a, b = get_value()
    wynik = slownik[oper](a, b)
    logger.info(f'Wynik dzialania: {wynik}')
    return wynik
````

````python
kalkulator()
````

#### Ćwiczenie dekoratory
zaimplementuj dekoratory bold i italic, które tekst zwracany przez inne funkcje będą otaczać znacznikami - odpowiednio <b>Oryginalny tekst</b> i <i>Oryginalny tekst</i>

@bold

zlacz_teksty("A", "B") == "<b>A\nB</b>

````python
import time
def bold(f):
    def wrapper(*args, **kwargs):
        text = f'<b>{f(*args, **kwargs)}</b>'
        return text
    return wrapper

def italic(f):
    def wrapper(*args, **kwargs):
        text = f'<i>{f(*args, **kwargs)}</i>'
        return text
    return wrapper

def timeit(f):
    def wrapper(*args, **kwargs):
        print(f'Wywołuję funkcję {f.__name__}')
        t = time.time()
        res = f(*args, **kwargs)
        print(f'Wywołanie zajęło {time.time() - t}')
        return res
    return wrapper
````

````python
@italic
@bold
def zlacz_teksty(*args, sep="\n"): 
    return "\n".join(args)
````

````python
zlacz_teksty('a', 'b', 'pies', 'kot')
````

````python
@timeit
def mil_sq():
    return [x ** 2 for x in range(1_000_000)]
````

````python
class Ksiazka:
    liczba_ksiazek = 0
    def __init__(self, tytul, autor, rok_wydania):
        self.tytul = tytul
        self.autor = autor
        self.rok_wydania = rok_wydania
        self.czy_wypozyczona = False
        self.__class__.liczba_ksiazek += 1
````

````python
import csv
class Ksiazka:
    _liczba_ksiazek = 0
    lista_instancji = []
    def __init__(self, tytul, autor, rok_wydania):
        self.tytul = tytul
        self.autor = autor
        self.rok_wydania = rok_wydania
        self._czy_wypozyczona = False
        self.increment()
        Ksiazka.lista_instancji.append(self)

    def wypozycz(self):
        if self._czy_wypozyczona:
            return 'Książka już jest wypożyczona'
        self._czy_wypozyczona = True

    def zwroc(self):
        if not self._czy_wypozyczona:
            return 'Książka nie jest wypożyczona.'
        self._czy_wypozyczona = False

    def info(self):
        czy_wyp = 'tak' if self._czy_wypozyczona else 'nie'
        return f'Tytuł: {self.tytul}, autor: {self.autor}, rok wydania: {self.rok_wydania}, wypożyczona: {czy_wyp}'
    
    @classmethod
    def increment(cls):
        cls._liczba_ksiazek += 1

    @classmethod
    def decrement(cls):
        cls._liczba_ksiazek -= 1

    @classmethod
    def ile_ksiazek(cls):
        return f'W systemie mamy {cls._liczba_ksiazek} książek'

    @classmethod
    def tworz_ksiazki(cls, plik):
        ksiazki = []
        with open(plik, newline='') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                ksiazki.append(Ksiazka(tytul=row['tytul'], autor=row['autor'], rok_wydania=row['rok_wydania']))
        return ksiazki

    @classmethod
    def check_counter(cls):
        return cls._liczba_ksiazek

    # def __del__(self):
    #     i = Ksiazka.lista_instancji.index(self)
    #     print(i)
    #     Ksiazka.lista_instancji.pop[i]
    #     Ksiazka.decrement()
````

#### Ćwiczenie property

````python
import math
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius
        
    @radius.setter
    def radius(self, value):
        if value <= 0:
            raise ValueError('Promień musi być większy od 0.')
        self._radius = value

    @property
    def area(self):
        return round(math.pi * self._radius ** 2, 2)

    @property
    def circumference(self):
        return round(2 * math.pi * self._radius, 2)

    @area.setter
    def area(self, value):
        if value <= 0:
            raise ValueError('Powierzchnia musi być większa od 0.')
        self._radius = math.sqrt(value / math.pi)

    @circumference.setter
    def circumference(self, value):
        if value <= 0:
            raise ValueError('Obwód musi być większy od 0.')
        self._radius = value / (2 * math.pi)
````
